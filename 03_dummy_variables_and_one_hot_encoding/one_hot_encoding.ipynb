```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
%matplotlib inline
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
le = LabelEncoder()
df.town = le.fit_transform(df.town)
df
```

```python
X = df[['town', 'area']].values
X
```

```python
y = df.price.values
y
```

```python
ohe = OneHotEncoder(sparse_output=False, dtype=int)
encoded_array = ohe.fit_transform(df[['town']])
encoded_array
```

```python
encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(['town']))
encoded_df
```

```python
df
```

```python
X = pd.concat([encoded_df, df[['area']]], axis='columns')
X
```

```python
X = X.drop(['town_0'],axis='columns')
X
```

```python
y = df.price
y
```

```python
model = LinearRegression()
model.fit(X,y)
```
