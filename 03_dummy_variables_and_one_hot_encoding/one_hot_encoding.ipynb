```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
%matplotlib inline
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
ohe = OneHotEncoder(sparse_output=False, dtype=int)
encoded_array = ohe.fit_transform(df[['town']])
encoded_array
```

```python
df_encoded = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(['town']))
df_encoded
```

```python
df_encoded
```

```python
X_train = pd.concat([df_encoded, df], axis='columns')
X_train
```

```python
X_train = X_train.drop(['town_monroe township', 'price', 'town'],axis='columns')
X_train
```

```python
y_train = df.price
y_train
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_train)
y_pred
```

```python
model.score(X_train,y_train)
```
